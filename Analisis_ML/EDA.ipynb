```python
#se cambian las columnas categoricas a numericas
from sklearn.preprocessing import OrdinalEncoder

encoder = OrdinalEncoder()

meta_train = pd.DataFrame(index = metadata.index)


for label in metadata.columns:
    print(label + "-encoded")
    #label = "Severity of Illness-encoded"
    encoder.fit(metadata[[label]])
    meta_train[label + "-encoded"] = encoder.transform(metadata[[label]])

display(metadata)
display(meta_train)
```

```python
def plot_corre_heatmap(corr):
    '''
    Definimos una función para ayudarnos a graficar un heatmap de correlación
    '''
    plt.figure(figsize=(12,10))
    sns.heatmap(corr, cbar = True,  square = False, annot=True, fmt= '.2f'
                ,annot_kws={'size': 15},cmap= 'coolwarm')
    plt.xticks(rotation = 45)
    plt.yticks(rotation = 45)
    # Arreglamos un pequeño problema de visualización
    b, t = plt.ylim() # discover the values for bottom and top
    b += 0.5 # Add 0.5 to the bottom
    t -= 0.5 # Subtract 0.5 from the top
    plt.ylim(b, t) # update the ylim(bottom, top) values
    plt.show()
```

```python
#Correlación
corr = meta_train.corr()
plot_corre_heatmap(corr)
```

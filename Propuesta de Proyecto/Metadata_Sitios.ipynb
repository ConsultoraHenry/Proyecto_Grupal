```python
#Crear un diccionario que contenga los límites de latitud y longitud para el estado de Florida. 
locator = Nominatim(user_agent="myGeocoder")
location = locator.geocode("Florida, USA")

lat_min, lat_max = location.raw['boundingbox'][0], location.raw['boundingbox'][1]
lon_min, lon_max = location.raw['boundingbox'][2], location.raw['boundingbox'][3]
```

```python
#pasar los datos a tipo float
lat_max = float(lat_max)
lat_min = float(lat_min)
lon_max = float(lon_max)
lon_min = float(lon_min)
```

```python
metadata_04 = metadata_04[(metadata_04['latitude'] >= lat_min) & (metadata_04['latitude'] <= lat_max) & 
             (metadata_04['longitude'] >= lon_min) & (metadata_04['longitude'] <= lon_max)]
```

```python
metadata_05 = metadata_05[(metadata_05['latitude'] >= lat_min) & (metadata_05['latitude'] <= lat_max) & 
             (metadata_05['longitude'] >= lon_min) & (metadata_05['longitude'] <= lon_max)]
```

```python
metadata_06 = metadata_06[(metadata_06['latitude'] >= lat_min) & (metadata_06['latitude'] <= lat_max) & 
             (metadata_06['longitude'] >= lon_min) & (metadata_06['longitude'] <= lon_max)]
```

Número de valores nulos para cada columna en el DataFrame.

text: es una columna que contiene el texto de la reseña. Hay un total de 1,079,510 valores faltantes en esta columna, lo que indica que algunos usuarios no han dejado ningún comentario en su reseña. Se considerara eliminar la columna.

pics: es una columna que indica si se incluyó alguna foto en la reseña. Hay un total de 2,745,810 valores faltantes en esta columna, lo que indica que la mayoría de las reseñas no incluyen fotos. Se considerara eliminar la columna.

resp: es una columna que indica si el negocio o alguien relacionado con él ha respondido a la reseña del usuario. Hay un total de 2,394,601 valores faltantes en esta columna, lo que indica que la mayoría de las reseñas no han recibido ninguna respuesta del negocio. Se considerara eliminar la columna.

```python
# Eliminar {} en la columna "resp"
reviews['resp'] = reviews['resp'].apply(lambda x: json.dumps(x))
reviews['resp'] = reviews['resp'].apply(lambda x: x.replace('{', '').replace('}', ''))
```

```python
# Reemplazar las listas por cadena de caracteres
reviews['user_id'] = reviews['user_id'].apply(lambda x: str(x).strip('[]'))
reviews['name'] = reviews['name'].apply(lambda x: str(x).strip('[]'))
reviews['time'] = reviews['time'].apply(lambda x: str(x).strip('[]'))
reviews['rating'] = reviews['rating'].apply(lambda x: str(x).strip('[]'))
reviews['text'] = reviews['text'].apply(lambda x: str(x).strip('[]'))
reviews['pics'] = reviews['pics'].apply(lambda x: str(x).strip('[]'))
reviews['resp'] = reviews['resp'].apply(lambda x: str(x).strip('[]'))
reviews['gmap_id'] = reviews['gmap_id'].apply(lambda x: str(x).strip('[]'))
```
